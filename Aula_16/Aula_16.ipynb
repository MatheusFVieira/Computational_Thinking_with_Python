```python
import pandas as pd

dados = pd.read_csv('/content/manutencao_preditiva.csv')
dados
```

```python
# Ex.1

df = pd.DataFrame({
    'A': [1, 2, 3, 4, 5],
    'B': [10, 20, 30, 40, 50],
    'C': [100, 200, 300, 400, 500]
})

df['A_dobro'] = df['A'].apply(lambda z: z*2)
df
```

```python
# Ex.2

df['soma_coluna'] = df.apply(lambda x: x['A'] + x['B'] + x['C'], axis=1)
df
```

```python
# Ex.3

import numpy as np

def calcular_media(linha):
  media = np.mean([linha['A'], linha['B'], linha['C']])
  return media

df['media_colunas'] = df.apply(calcular_media, axis=1)
df
```

```python
# Ex.4

df['categoria'] = df['A'].apply(lambda x: 'Alto' if x > 3 else 'Baixo')
df
```

```python
# Exercício 1

df['A_normalizado'] = (df['A'] - df['A'].min()) / (df['A'].max() - df['A'].min())
df['B_normalizado'] = (df['B'] - df['B'].min()) / (df['B'].max() - df['B'].min())
df['C_normalizado'] = (df['C'] - df['C'].min()) / (df['C'].max() - df['C'].min())
df
```

```python
# Exercício 2

dados['Temperatura_AR_N'] = (dados['Temperatura Ar [K]'] - dados['Temperatura Ar [K]'].min()) / (dados['Temperatura Ar [K]'].max() - dados['Temperatura Ar [K]'].min())
dados['Temperatura_Processo_N'] = (dados['Temperatura Processo [K]'] - dados['Temperatura Processo [K]'].min()) / (dados['Temperatura Processo [K]'].max() - dados['Temperatura Processo [K]'].min())
dados['Velocidade_Rotação_N'] = (dados['Velocidade Rotacao [rpm]'] - dados['Velocidade Rotacao [rpm]'].min()) / (dados['Velocidade Rotacao [rpm]'].max() - dados['Velocidade Rotacao [rpm]'].min())
dados['Torque_N'] = (dados['Torque [Nm]'] - dados['Torque [Nm]'].min()) / (dados['Torque [Nm]'].max() - dados['Torque [Nm]'].min())
dados
```

```python
# Plot das variáveis temperatura do Ar e Processo

from matplotlib import pyplot as plt

x = dados['UDI']
y1 = dados['Temperatura Ar [K]']
y2 = dados['Temperatura Processo [K]']

plt.plot(x, y1, x, y2)
plt.show()
```

```python
x = dados['UDI']
y1 = dados['Temperatura_AR_N']
y2 = dados['Temperatura_Processo_N']

plt.plot(x, y1, x, y2)
plt.show()
```
