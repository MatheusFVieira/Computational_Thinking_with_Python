```python
import pandas as pd

dados = pd.read_csv('/content/manutencao_preditiva.csv')

falha = {
    'UDI': [],
    'Temperatura Ar [K]': [],
    'Temperatura Processo [K]': [],
    'Tipo da Falha': []
}

for index, row in dados.iterrows():
    if row['Tipo da Falha'] != 'No Failure':
        falha['UDI'].append(row['UDI'])
        falha['Temperatura Ar [K]'].append(row['Temperatura Ar [K]'])
        falha['Temperatura Processo [K]'].append(row['Temperatura Processo [K]'])
        falha['Tipo da Falha'].append(row['Tipo da Falha'])


print(falha)
```

```python
# Exercicio 5

temp_ar = dados['Temperatura Ar [K]']
temp_processo = dados['Temperatura Processo [K]']

for i in range(len(temp_ar)):
    sum = temp_processo[i] - temp_ar[i]
    media = sum / len(temp_ar)

print(media)
```

```python
# Exercicio 6

produtos = dados['Tipo']

tipo_M = []

for i in range(len(produtos)):
    if produtos[i] == 'M':
        tipo_M.append(i)

print(len(tipo_M))
```

```python
# Exercicio 7

id = []

for index, row in dados.iterrows():
    if row['Velocidade Rotacao [rpm]'] > 1500 and row['Torque [Nm]'] > 40:
        id.append(row['ID Produto'])

print(id)
```

```python
# Exercicio 8

dados['eficiencia'] = dados['Torque [Nm]'] / dados['Velocidade Rotacao [rpm]']
media = dados['eficiencia'].mean()
print(media)
```

```python
# Exercicio 9

temperatura_media_por_falha = dados.groupby('Tipo da Falha')['Temperatura Processo [K]'].mean()

print("Temperatura média do processo por tipo de falha:")
print(temperatura_media_por_falha)

maior_temperatura_media = temperatura_media_por_falha.idxmax()
print(f"\nO tipo de falha com a maior temperatura média é: {maior_temperatura_media}")
```
