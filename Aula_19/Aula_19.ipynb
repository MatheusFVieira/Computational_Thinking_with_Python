```python
import pandas as pd

dados = {
    'produto': ['Notebook', 'Mouse', 'Teclado', 'Minitor', 'Webcam'],
    'vendas': [1200, 300, 450, 800, None],
    'lucro': [300, 50, 80, 200, 40],
    'ano': ['2022', '2022', '2022', '2022', 'dois mil e vinte e dois']
}

df = pd.DataFrame(dados)
```

```python
try:
  media = df['vendas'].mean()
  print(f"Média do estoque: {media}")
except KeyError:
  print("A coluna de 'estoque' não existe no DataFrame.")
```

```python
try:
  media_vendas = df['vendas'].mean()
  print(f"Média de vendas: {media_vendas:.2f}")
except Exception as e:
  print(f"Erro ao calcular a média de vendas. {e}")
```

```python
try:
  df["margem"] = df["lucro"] / df["vendas"]
except ZeroDivisionError:
  print("Erro: Divisão por zero.")
except Exception as e:
  print(f"Erro: {e}")
else:
  print("Operação realizada com sucesso.")
finally:
  print("Fim da operação.")
df
```

```python
df.to_csv("dados.csv")
df.to_excel("dados.xlsx")
df.to_json("dados.json", orient='records', indent=2, force_ascii=False)
```
