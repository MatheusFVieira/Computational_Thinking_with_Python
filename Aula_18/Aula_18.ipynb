```python
# Ex.1

class Pessoa:
  def __init__(self, nome, idade):
    self.nome = nome
    self.idade = idade

  def apresentar(self):
    print(f"Olá! Meu nome é {self.nome} e tenho {self.idade} anos.")

p1 = Pessoa("Matheus", 19)
p2 = Pessoa("Maria", 30)

p1.apresentar()
p2.apresentar()
```

```python
# Ex.2

pessoa = [
    Pessoa('Ana', 25),
    Pessoa('João', 30),
    Pessoa('Maria', 28)
]

for p in pessoa:
  p.apresentar()
```

```python
# Ex.3

class Informações:
  def __init__(self, nacionalidade, idade, estado_civil):
    self.nacionalidade = nacionalidade
    self.idade = idade
    self.estado_civil = estado_civil

  def apresentar(self):
    print(f"Sou {self.nacionalidade} e tenho {self.idade} anos. Sou {self.estado_civil}.")

i1 = Informações('Brasileiro', 19, 'Solteiro')
i2 = Informações('Brasileiro', 30, 'Casado')

i1.apresentar()
i2.apresentar()
```

```python
# Ex.4

class poligono:
  def __init__(self, base, altura):
    self.base = base
    self.altura = altura

  def calcular_perimetro(self):
    perimetro = 2 * (self.base + self.altura)
    return perimetro

  def calcular_area(self):
    area = self.base * self.altura
    return area

  def apresentar(self):
    print(f"O perímetro do polígono é {self.calcular_perimetro()} e a área é {self.calcular_area()}.")

p1 = poligono(5, 10)
p2 = poligono(25, 32)

p1.apresentar()
p2.apresentar()
```

```python
# Ex.5

class Padroniza:
  def __init__(self, usuario, dominio):
    self.usuario = usuario
    self.dominio = dominio

  def formatar(self):
    print(f"{self.usuario}@{self.dominio}")

u1 = Padroniza(input("Digite o nome de usuário: "), input("Digite o domínio: "))
u1.formatar()
```

```python
# Ex.6

import pandas as pd

class imc:
  def __init__(self, peso, altura):
    self.peso = peso
    self.altura = altura

  def calcular_imc(self):
    imc = self.peso / (self.altura ** 2)
    return imc

pessoa = [
    imc(60, 160),
    imc(80, 185),
    imc(70, 180)
]

dataFrame = pd.DataFrame([p.calcular_imc() for p in pessoa])
dataFrame
```

```python
# Ex.7

class Juros:
  def __init__(self, valor, taxa):
    self.valor = valor
    self.taxa = taxa

  def calcular_juros(self):
    juros = self.valor * (1 + (self.taxa / 100)**2)
    return juros

contas = [
    Juros(1000, 10),
    Juros(500, 8),
    Juros(1500, 11)
]

df = pd.DataFrame([c.calcular_juros() for c in contas])
df
```
