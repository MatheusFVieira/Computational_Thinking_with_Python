```python
def acessar_lista():
  lista = [10, 20, 30, 40 , 50]

  try:
    indice = int(input("Digite o índice da lista: "))
    print(f"O valor no índice {indice} é {lista[indice]}")
  except IndexError:
    print("Índice inválido")
  except ValueError:
    print("Você não digitou um número inteiro")

acessar_lista()
```

```python
# Exercício 1

def converter_temperatura():
  try:
    temperatura = float(input("Digite a temperatura em graus Celsius: "))
    fahrenheit = (temperatura * 9/5) + 32
    print(f"A temperatura em Fahrenheit é {fahrenheit}°F")
  except ValueError:
    print("Insira um valor válido")

converter_temperatura()
```

```python
# Exercício 2

def calcular_area():
  try:
    escolha = int(input("Digite 1 para calcular a área de um retangulo ou 2 para calcular a área de um círculo: "))
    if escolha == 1:
      base = float(input("Digite a base do retângulo: "))
      altura = float(input("Digite a altura do retângulo: "))
      area = base * altura
      print(f"A área do retângulo é {area}")
    elif escolha == 2:
      raio = float(input("Digite o raio do círculo: "))
      if raio <=0:
        raise ValueError("O raio não é um valor válido!")
      area = 3.1415 * (raio ** 2)
      print(f"A área do círculo é {area}")
  except ValueError:
    print("Insira um valor válido")

calcular_area()
```
